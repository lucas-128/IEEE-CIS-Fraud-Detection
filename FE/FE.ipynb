<strong>Features de tiempo (transaction_dt)</strong></p> 

```python
train_set['Hora'] = train_set.transaction_dt/3600
train_set.Hora = train_set.Hora.astype(int)
train_set.Hora = train_set.Hora % 24
```

```python
train_set['numDia'] =  train_set.transaction_dt/86400
train_set.numDia = train_set.numDia.astype(int)
train_set['diaSemana'] =  train_set.numDia % 7
```

```python
train_set.drop(['transaction_dt','numDia'], axis=1,inplace = True)
```

```python
test_set['Hora'] = test_set.transaction_dt/3600
test_set.Hora = test_set.Hora.astype(int)
test_set.Hora = test_set.Hora % 24
test_set['numDia'] =  test_set.transaction_dt/86400
test_set.numDia = test_set.numDia.astype(int)
test_set['diaSemana'] =  test_set.numDia % 7
test_set.drop(['transaction_dt','numDia'], axis=1,inplace = True)
```

<a id = "1.0"></a>
<p style="font-size:18px;color:black;font-family:'Arial';text-align:left;border-radius:5px 5px 5px 5px;">
<strong>Resolucion del dispositivo</strong></p> 

```python
train_set['id_33'].head(10)
```

```python
def multiplicarResolucion(texto):
  if(texto != 'None'):
    lista = texto.split('x')
    return int(lista[0]) * int(lista[1])
  return np.nan
```

```python
test_set['id_33'] = test_set['id_33'].fillna('None')
test_set['pixelCount'] =  test_set['id_33'].apply(multiplicarResolucion)
test_set.drop(['id_33'], axis=1,inplace = True)
```

```python
train_set['id_33'] = train_set['id_33'].fillna('None')
train_set['pixelCount'] =  train_set['id_33'].apply(multiplicarResolucion)
train_set.drop(['id_33'], axis=1,inplace = True)
```

```python
train_set['pixelCount'].head()
```

```python
train_set.drop(['transaction_id'], axis=1,inplace = True)
test_set.drop(['transaction_id'], axis=1,inplace = True)
```

<a id = "1.0"></a>
<p style="font-size:18px;color:black;font-family:'Arial';text-align:left;border-radius:5px 5px 5px 5px;">
<strong>Mean encoding</strong></p> 

```python
diccionarios = {}
columnasMeanEncodeadas = []

for columna in train_set.columns:
  if (train_set[columna].dtype==object):
    Mean_encoded_subject = train_set.groupby([columna])['is_fraud'].mean().to_dict()
    diccionarios[columna] = Mean_encoded_subject
    columnasMeanEncodeadas.append(columna)
    train_set[columna] = train_set[columna].map(Mean_encoded_subject)
```

```python
columnasMeanEncodeadas
```

```python
for columna in columnasMeanEncodeadas:
  test_set[columna] = test_set[columna].map(diccionarios[columna])
```

```python
# Devuelve True si hay un elemento menor a 0, false caso contrario 
train_set.lt(0).any()
```

```python
# Si la columna no tiene elementos menores a 0, reemplazo con -1; si no, reemplazo con la media
for col in train_set.columns:
    if not train_set[col].lt(0).any():
        train_set.fillna(-1,inplace=True)
    else:
        train_set.fillna(train_set[col].mean(),inplace=True)
```

```python
for col in test_set.columns:
    if not test_set[col].lt(0).any():
        test_set.fillna(-1,inplace=True)
    else:
        test_set.fillna(test_set[col].mean(),inplace=True)
```

<strong>PCA</strong></p> 

```python
from sklearn.decomposition import PCA
```

```python
train_set.head()
```

```python
v_columns = [column for column in train_set.columns if column.startswith('v')]
```

```python
v_columns
```

```python
pca = PCA(n_components=20)
principalComponents = pca.fit_transform(train_set[v_columns])
principalDf = pd.DataFrame(principalComponents)
train_set.drop(v_columns, axis=1, inplace=True)
principalDf.rename(columns=lambda x: str("V_")+str(x), inplace=True)
train_set = pd.concat([train_set, principalDf], axis=1)
```

```python
train_set.head()
```

```python
train_set.shape
```

```python
train_set.info()
```

```python
pca = PCA(n_components=20)
principalComponents = pca.fit_transform(test_set[v_columns])
principalDf = pd.DataFrame(principalComponents)
test_set.drop(v_columns, axis=1, inplace=True)
principalDf.rename(columns=lambda x: str("V_")+str(x), inplace=True)
test_set = pd.concat([test_set, principalDf], axis=1)
```
