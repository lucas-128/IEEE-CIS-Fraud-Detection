```python
for col in test_identity.columns:
    if col.startswith('id-'):
        partes = col.split('-')
        test_identity['id_' + partes[1]] = test_identity[col]
        test_identity.drop(col, inplace=True, axis=1)
```

Mean encoding + Fill NaNs

```python
transaction.dtypes
```

```python
for columna in transaction.columns:
  if (transaction[columna].dtype==object):
    Mean_encoded_subject = transaction.groupby([columna])['isFraud'].mean().to_dict()
    transaction[columna] = transaction[columna].map(Mean_encoded_subject)
```

```python
identity.dtypes
```

```python
id_isFraud = transaction[['TransactionID', 'isFraud']]
```

```python
id_isFraud.head()
```

```python
identity2 = pd.merge(identity, id_isFraud, on='TransactionID')
```

```python
for columna in identity2.columns:
  if (identity2[columna].dtype==object):
    Mean_encoded_subject = identity2.groupby([columna])['isFraud'].mean().to_dict()
    identity2[columna] = identity2[columna].map(Mean_encoded_subject)
```

```python
for col in transaction.columns:
    if not transaction[col].lt(0).any():
        transaction.fillna(-1,inplace=True)
    else:
        transaction.fillna(transaction[col].mean(),inplace=True)
```

```python
for col in identity2.columns:
    if not identity2[col].lt(0).any():
        identity2.fillna(-1,inplace=True)
    else:
        identity2.fillna(identity2[col].mean(),inplace=True)
```

```python
print(transaction.isna().sum().sum())
print(identity2.isna().sum().sum())
```

FeatureTools

```python
import featuretools as ft

# En produccion eliminar, se incluye por informacion
print(f'Featuretools Version: {ft.__version__}') 
```

```python
entities = {
        "identity" : (identity2, "TransactionID"),
        "transaction" : (transaction, "TransactionID")
}
```

```python
es = ft.EntitySet("my-entity-set", entities)
```

```python
feature_matrix_customers, features_defs = ft.dfs(
    dataframes=entities,
    #relationships=relationships,
    target_dataframe_name="identity",
)
```
