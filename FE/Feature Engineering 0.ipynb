```python
#!pip install klib
import klib
```

```python
train_clean = klib.clean_column_names(train_set)
test_clean = klib.clean_column_names(test_set)
```

```python
train_clean = klib.convert_datatypes(train_clean)
test_clean = klib.convert_datatypes(test_clean)
```

```python
test_clean.info()
```

```python
train_clean.info()
```

```python
train_clean['Hora'] = train_clean.transaction_dt/3600
train_clean.Hora = train_clean.Hora.astype(int)
train_clean.Hora = train_clean.Hora % 24
```

```python
train_clean['numDia'] =  train_clean.transaction_dt/86400
train_clean.numDia = train_clean.numDia.astype(int)
train_clean['diaSemana'] =  train_clean.numDia % 7
```

```python
train_clean.drop(['transaction_dt','numDia'], axis=1,inplace = True)
```

```python
test_clean['Hora'] = test_clean.transaction_dt/3600
test_clean.Hora = test_clean.Hora.astype(int)
test_clean.Hora = test_clean.Hora % 24
test_clean['numDia'] =  test_clean.transaction_dt/86400
test_clean.numDia = test_clean.numDia.astype(int)
test_clean['diaSemana'] =  test_clean.numDia % 7
test_clean.drop(['transaction_dt','numDia'], axis=1,inplace = True)
```

```python
train_clean['id_33'].head(10)
```

```python
cat_cols = train_clean.select_dtypes(include='category').columns
train_clean[cat_cols] = train_clean[cat_cols].astype('object')
```

```python
cat_cols = test_clean.select_dtypes(include='category').columns
test_clean[cat_cols] = test_clean[cat_cols].astype('object')
```

```python
def multiplicarResolucion(texto):
  if(texto != 'None'):
    lista = texto.split('x')
    return int(lista[0]) * int(lista[1])
  return np.nan
```

```python
test_clean['id_33'] = test_clean['id_33'].fillna('None')
test_clean['pixelCount'] =  test_clean['id_33'].apply(multiplicarResolucion)
test_clean.drop(['id_33'], axis=1,inplace = True)
```

```python
train_clean['id_33'] = train_clean['id_33'].fillna('None')
train_clean['pixelCount'] =  train_clean['id_33'].apply(multiplicarResolucion)
train_clean.drop(['id_33'], axis=1,inplace = True)
```

```python
train_clean['pixelCount'].head()
```

```python
train_clean.drop(['transaction_id'], axis=1,inplace = True)
test_clean.drop(['transaction_id'], axis=1,inplace = True)
```

<p style="font-size:18px;color:black;font-family:'Arial';text-align:left;border-radius:5px 5px 5px 5px;">
<strong>Mean encoding</strong></p> 

```python
diccionarios = {}
columnasMeanEncodeadas = []

for columna in train_clean.columns:
  if (train_clean[columna].dtype==object):
    Mean_encoded_subject = train_clean.groupby([columna])['is_fraud'].mean().to_dict()
    diccionarios[columna] = Mean_encoded_subject
    columnasMeanEncodeadas.append(columna)
    train_clean[columna] = train_clean[columna].map(Mean_encoded_subject)
```

```python
columnasMeanEncodeadas
```

```python
train_clean.info()
```

```python
for columna in columnasMeanEncodeadas:
  test_clean[columna] = test_clean[columna].map(diccionarios[columna])
```

```python
test_clean.info()
```

```python
for col in train_clean.columns:
    if not train_clean[col].lt(0).any():
        train_clean.fillna(-1,inplace=True)
    else:
        train_clean.fillna(train_clean[col].mean(),inplace=True)
```

```python
for col in test_clean.columns:
    if not test_clean[col].lt(0).any():
        test_clean.fillna(-1,inplace=True)
    else:
        test_clean.fillna(test_clean[col].mean(),inplace=True)
```
