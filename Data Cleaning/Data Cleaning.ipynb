```python
def get_columns_not_in_other_df(df1, df2):
  """Returns the names of the columns which are in df1 but not in df2."""
  columns_in_df1 = set(df1.columns)
  columns_in_df2 = set(df2.columns)
  columns_not_in_df2 = list(columns_in_df1 - columns_in_df2)
  return columns_not_in_df2
```
